```python

import cv2 
import matplotlib.pylab as plt

img = cv2.imread('image/day.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)


hsv[:,:,2] =hsv[:,:,2]/2 
img2 = hsv
img2 = cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)

plt.figure(figsize=(16,16))
plt.subplot("121")
plt.imshow(img)
plt.subplot("122")
plt.imshow(img2)
```
